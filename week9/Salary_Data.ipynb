```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd
# Import the stats library from numpy
from scipy import stats
# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
# Command so that plots appear in the iPython Notebook
%matplotlib inline

```

```python
#Read f
salary_Data= pd.read_csv("Salary_Data.csv")
```

```python
sns.scatterplot(x='Salary',y='Years of Experience',data=salary_Data)
```

```python
salary_Data=salary_Data.dropna()
```

```python
sns.regplot(x='Salary',y='Years of Experience',data=salary_Data)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(salary_Data['Salary'],salary_Data['Years of Experience'])
print("y="+str(np.round(slope,2))+"x+"+str(np.round(intercept,2)))
```

```python
#R square in a regression analysis is called the correlation coefficient 



slope, intercept, r_value, p_value, std_err
```

```python


stats.pearsonr(salary_Data['Salary'],salary_Data['Years of Experience'])
```
