```python

# Let's begin by importing what we'll need (You'll probably be copying and␣→pasting this a lot)
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd
# Import the stats librayr from numpy
from scipy import stats
# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

```python
#Create a random normal-dist dataset
dataset1 = randn(100)
#Plot a histogram of the dataset, note bins=10 by default
plt.hist(dataset1)
```

```python
dataset1
```

```python
# Lets make another dataset
dataset2 = randn(80)
#Plot
plt.hist(dataset2,color='indianred',edgecolor='black')
```

```python
dataset2
```

```python
# We can use normed to plot on same plot
# Set normed/density=True for the plots to be normalized in order to comapre␣ data sets with different number of observations
# Set alpha=0.5 for transperancy
plt.hist(dataset1,density=True,color='indianred',alpha=0.5,bins=20)
plt.hist(dataset2,density=True,alpha=0.5,bins=20)
```

```python
# Make two more random normal dist data sets randn is normal distribution
data1 = randn(1000,)
data2 = randn(1000)
#Can represent joint distributions using joint plots


data=pd.DataFrame(data1,data2)
sns.jointplot(x=data1,y=data2)
```

```python
data1
```

```python
data2
```

```python
# Can also use hex bins for a more concise picture
#kind=scatter | reg | resid | kde | hex
p=sns.jointplot(x=data1,y=data2,kind='kde')
p.set_axis_labels("Data1", "Data2")
```

```python
#more advanced options
plt.hist(x=data1, bins=10, color='#0504aa',alpha=0.5, rwidth=0.8)
plt.grid(axis='y', alpha=0.75) #'both'
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('My Very Own Histogram')
plt.savefig("hist.jpeg",transparent=True)
```

```python
#Kernal Desnity Destribution to estimate histogram
#http://en.wikipedia.org/wiki/Kernel_density_estimation
sns.distplot(data1)
```

```python
#Kernal Desnity Destribution to estimate histogram
#http://en.wikipedia.org/wiki/Kernel_density_estimation
sns.kdeplot(data1)

#Kernal Desnity Destribution
sns.kdeplot(data1)
#plot different KDD with different bandwidth "bw" and shaded
for i in np.arange(0.5,3,0.5):
    sns.kdeplot(data1,bw=i,label=i,shade=True)
```

```python
#Vertical drawing
sns.kdeplot(data1, vertical=True)
```

```python
# Finally we can also use kde plot to create a cumulative distribution function␣↪→(CDF) of the data
# URL for info on CDF
url = 'http://en.wikipedia.org/wiki/Cumulative_distribution_function'
sns.kdeplot(data1,cumulative=True)
```
