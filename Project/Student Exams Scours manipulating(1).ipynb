# 2) violinplot:
The violin plots in the code you provided offer a visualization of the distribution of scores for Math, Reading, and Writing subjects in the dataset. They provide insights into the shape, central tendency, and density of the score distributions

* Width: The width of the violin at a specific point represents the density or frequency of scores at that value. A wider section indicates a higher density of scores, while a narrower section indicates a lower density.

* Median: The white dot or line 

```python
plt.figure(figsize=(15,5))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(141)
plt.title('Math Scores')
sns.violinplot(y='MathScore',data=student_frame,color='m',linewidth=2)
plt.subplot(142)
plt.title('Reading Scores')
sns.violinplot(y='ReadingScore',data=student_frame,color='g',linewidth=2)
plt.subplot(143)
plt.title('Writing Scores')
sns.violinplot(y='WritingScore',data=student_frame,color='r',linewidth=2)
plt.show()

#Refference
url='https://www.kaggle.com/code/amulyamanne/eda-on-students-performance-in-exams/notebook'
```

# 3) Boxplot(Scors ,LunchType,TestPrep):

The code you provided generates box plots to visualize the relationship between the(ReadingScore, WritingScore, MathScore) and LunchType, with an additional differentiation based on TestPrep.

The box plots are created separately for each combination of TestPrep and Gender. The LunchType is represented on the x-axis, and the (ReadingScore, WritingScore, and MathScore) are displayed on the y-axis. The boxes in the plot represent the interquartile range (IQR), with the median marked as a line inside the box. Any points outside this range are considered outliers and displayed as individual dots.

By using different colors for each gender, the plots show how the distribution of (ReadingScore, WritingScore, and MathScore) varies across different LunchTypes, taking into account the TestPrep variable.



```python
plt.figure(figsize=(8,7))
sns.catplot(y="WritingScore", x="LunchType", hue="Gender",col='TestPrep',kind='box',
data=student_frame, palette="Set3")

plt.figure(figsize=(8,7))
sns.catplot(y="ReadingScore", x="LunchType", hue="Gender",col='TestPrep',kind='box',
data=student_frame, palette="Set3")

plt.figure(figsize=(8,7))
sns.catplot(y="MathScore", x="LunchType", hue="Gender",col='TestPrep',kind='box',
data=student_frame, palette="Set3")
```

Analyzing the box plots can provide insights into the relationship between (WritingScore, ReadingScore, and MathScore) in LunchType and TestPrep. You can observe the following aspects:

We note that female scores are higher than male scores in Reading and Writing, whether they were completed test preparation course or not, and regardless of the lunch type they got. On the other hand, male students score is higher than the female in Math exam.

# Regression Plots:
* Regression plots are a useful tool for visualizing the relationship between two variables and estimating a linear regression model. In this code, I am creating a regression plot to explore the relationship between (WritingScore, MathScore, Reading Score) and "EthnicGroup" in the dataset.

```python
slope, intercept, r_value, p_value, std_err
```

1. slope: The slope represents the rate of change in the dependent variable (y in y = mx + b) with respect to a one-unit change in the independent variable (x)
2. intercept:  the value of the dependent variable when the independent variable is zero.
3. r_value:  also known as the correlation coefficient or Pearson's correlation coefficient
4. p_value: is a measure of the statistical significance of the estimated relationship between the variables. It indicates the probability of obtaining a relationship as strong as or stronger than the observed relationship by chance alone
5. std_err: The `std_err`, or standard error, quantifies the average amount of variability or dispersion in the observed values around the estimated regression line

```python
#Linear Regression line between MathScore and WritingScore
#We can also specify the confidence interval to use for the linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(student_frame['WritingScore'],student_frame['MathScore'])
print("y="+str(np.round(slope,2))+"x+"+str(np.round(intercept,2)))
```

```python
#Linear Regression line between ReadingScore and WritingScore
#We can also specify the confidence interval to use for the linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(student_frame['WritingScore'],student_frame['ReadingScore'])
print("y="+str(np.round(slope,2))+"x+"+str(np.round(intercept,2)))
```

```python
#Linear Regression line between MathScore and ReadingScore
#We can also specify the confidence interval to use for the linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(student_frame['MathScore'],student_frame['ReadingScore'])
print("y="+str(np.round(slope,2))+"x+"+str(np.round(intercept,2)))
```
